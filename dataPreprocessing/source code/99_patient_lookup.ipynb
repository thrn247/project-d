### 1. SETUP PATH TO KEY FILE
This chunk executes the relevant logic for the step mentioned above.

```python
import pandas as pd
import os

# 1. SETUP PATH TO KEY FILE
# Assumes this notebook is in the same folder as your other steps
current_dir = os.getcwd()
key_file = os.path.join(current_dir, "..", "csv", "patient_id_key.xlsx")

# 2. PERFORM LOOKUP
if os.path.exists(key_file):
    df_key = pd.read_excel(key_file)
    print(" Patient Key Loaded successfully.")
    
    # Ask for Input
    target_id = input("Enter Patient ID (e.g., P00054): ").strip()
    
    # Find the row
    result = df_key[df_key["Patient_ID"] == target_id]
    
    if not result.empty:
        real_code = result.iloc[0]["PATIENT_CODE"]
        print("-" * 30)
        print(f"🔍 SEARCH RESULT")
        print(f"🆔 Anonymized ID: {target_id}")
        print(f"👤 REAL Identity: {real_code}")
        print("-" * 30)
    else:
        print(f" ID '{target_id}' not found in the key file.")
else:
    print(" Error: 'patient_id_key.xlsx' not found.")
    print("   Did you run Step 4 (Aggregation) to generate it?")
```
