### 1. Define Paths Based On Your Main Rig'S Structure

This chunk executes the relevant logic for the step mentioned above.

```python
import pandas as pd
import numpy as np
import os

# 1. Define Paths based on your main rig's structure
base_path = r"C:\Users\thiranbarath\Documents\GitHub\project-d"
input_path = os.path.join(base_path, "dataPreprocessing", "csv", "step4_patient_aggregated.xlsx")

print(f"Looking for data at: {input_path}")

# 2. Load the dataset
df = pd.read_excel(input_path)

print(f"✅ Data loaded successfully!")
print(f"Dataset Shape (Rows, Columns): {df.shape}")

# Peek at the first 3 rows to confirm
df.head(3)
```

### 1. Check What The Current Text Values Look Like

This chunk executes the relevant logic for the step mentioned above.

```python
# 1. Check what the current text values look like
if 'SEVERITY_INDEX' in df.columns:
    print(f"Original Severity categories: {df['SEVERITY_INDEX'].unique()}")
    
    # 2. Create the mapping dictionary
    severity_mapping = {
        'Mild': 0,
        'Moderate': 1,
        'Severe': 2
    }
    
    # 3. Apply mapping and create a new numeric column
    df['Severity_Encoded'] = df['SEVERITY_INDEX'].map(severity_mapping).fillna(0) # 0 = Mild as safe fallback
    
    # 4. Drop the old text column
    df.drop('SEVERITY_INDEX', axis=1, inplace=True)
    
    print("✅ 'SEVERITY_INDEX' successfully converted to numbers (0, 1, 2) and original column dropped.")
else:
    print("⚠️ 'SEVERITY_INDEX' column not found. It might have already been processed.")

# Verify the change
df['Severity_Encoded'].value_counts(dropna=False)
```

### 1. Drop Only Patient_Id

This chunk executes the relevant logic for the step mentioned above.

```python
# 1. Drop only Patient_ID
columns_to_drop = ['Patient_ID']
df_ml = df.drop(columns=columns_to_drop, errors='ignore')
print(f"Dropped columns: {columns_to_drop}")

# 2. One-Hot Encoding for 'SEX'
# Added `dtype=int` to force it to output 1 and 0 instead of True/False
if 'SEX' in df_ml.columns:
    df_ml = pd.get_dummies(df_ml, columns=['SEX'], drop_first=False, dtype=int)
    print("✅ Categorical variable 'SEX' successfully One-Hot Encoded (as 1s and 0s).")

# 3. Check the final shape and data types
print(f"\nFinal ML Dataset Shape: {df_ml.shape}")
print("\nChecking data types:")
print(df_ml.dtypes)
```

### 1. Define The Path For The New Machinelearning Folder

This chunk executes the relevant logic for the step mentioned above.

```python
import os

# 1. Define the path for the new machineLearning folder
# Using the same base_path we defined in Cell 1
output_dir = os.path.join(base_path, "machineLearning", "csv")
output_file = os.path.join(output_dir, "ml_ready_dataset.csv")

# 2. Create the folder if it doesn't exist yet
os.makedirs(output_dir, exist_ok=True)

# 3. Save the dataframe to CSV
# We use index=False so pandas doesn't write row numbers as a new column
df_ml.to_csv(output_file, index=False)

print(f"✅ Master ML Dataset successfully saved to:")
print(output_file)
```
